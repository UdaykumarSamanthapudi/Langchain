```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import  HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
vector=embedding.embed_query("I love programming")
print(vector)
```

```python
from langchain_community.document_loaders import  PyPDFLoader

loader=PyPDFLoader("resources/sample.pdf").load()
texts = [doc.page_content for doc in loader]


pdfEmbedding=embedding.embed_documents(texts)
print(pdfEmbedding)
print("length of embeddings: ",len(pdfEmbedding))
print("vector dimensions: ",len(pdfEmbedding[0]))
```
