```python
#USING SIMPLE CHAIN
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

llm=ChatGroq(model="Llama-3.1-8B-instant")
prompt=ChatPromptTemplate.from_template("what is {topic}")

chain= prompt | llm
chain.invoke({"topic":"java"})
```

```python
#Using Sequential Chain
from langchain_core.output_parsers import StrOutputParser
prompt1=ChatPromptTemplate.from_template("Give me about {topic}")
chain1= prompt1 | llm | StrOutputParser()

res=chain1.invoke({"topic":"Java"})

prompt2=ChatPromptTemplate.from_template("What is the topic of this {title}")
chain2 = prompt2 | llm | StrOutputParser()
result=chain2.invoke({"title":res})
print(result)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

# Model
llm=ChatGroq(model="Llama-3.1-8B-instant")

# Step 1: Joke prompt
joke_prompt = ChatPromptTemplate.from_template(
    "Tell me a short joke about {topic}"
)

# Joke chain
joke_chain = joke_prompt | llm | StrOutputParser()

# Step 2: Explanation prompt
explain_prompt = ChatPromptTemplate.from_template(
    "Explain why this joke about {topic} is funny:\n{joke}"
)

# Full chain
chain = (
    {
        "topic": RunnablePassthrough(),
        "joke": joke_chain
    }
    | explain_prompt
    | llm
    | StrOutputParser()
)

# Run the chain
result = chain.invoke({"topic": "programmers"})

print(result)
```

```python
#Using Parallel Chains
from langchain_core.runnables import RunnableParallel
prompt1=ChatPromptTemplate.from_template("what is {topic}")
chain1=prompt1 | llm | StrOutputParser()

prompt2=ChatPromptTemplate.from_template("when was  {topic} created ?")
chain2=prompt2 | llm | StrOutputParser()

prompt3=ChatPromptTemplate.from_template("who invented  {topic}")
chain3=prompt3 | llm | StrOutputParser()

parallelChain=RunnableParallel(
    about=chain1,
    created=chain2,
    inventedBy=chain3
)

res=parallelChain.invoke({"topic":"Car"})
print(res)
```
