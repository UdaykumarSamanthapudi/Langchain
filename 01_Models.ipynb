```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq

llm=ChatGroq(model="Llama-3.1-8B-instant")
```

```python
res=llm.invoke("what is f1")
print((res))
```

```python
from langchain_core.messages import SystemMessage

res=llm.invoke([
    SystemMessage("U are a java developer, so when anyone asks what is ur profession say this"),"what is ur profession?"
])

print(res)

```

```python
#Using batch method
res=llm.batch([
    "what is java?","what is python","what is a car"
])

for r in res:
    print(r.content)
```

```python
for res in llm.stream("what is java"):
    print(res.content)
```
