```python
#Using FAISS DB
from langchain_community.vectorstores import  FAISS
from langchain_community.document_loaders import  TextLoader,PyPDFLoader
from langchain_text_splitters import  RecursiveCharacterTextSplitter
from langchain_huggingface import  HuggingFaceEmbeddings

loader=PyPDFLoader("resources/Java.pdf")
docs=loader.load()

splitter=RecursiveCharacterTextSplitter(chunk_size=400,chunk_overlap=100)
documents=splitter.split_documents(docs)

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

doc_embeddings=embedding.embed_documents(documents)
vector_db=FAISS.from_documents(documents=documents,embedding=embedding)

similarity=vector_db.similarity_search("what is Java",k=3)
for s in similarity:
    print(s.page_content)
print(similarity[0].page_content)
```
