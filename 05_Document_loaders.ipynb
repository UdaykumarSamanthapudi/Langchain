```python
#Using Directory Loader
from langchain_community.document_loaders import DirectoryLoader, TextLoader

loader = DirectoryLoader(
    "resources",
    glob="**/*.txt",
    loader_cls=TextLoader
)

docs = loader.load()
print(docs)
```

```python
#Using Directory Loader to load all kinds of files
from langchain_community.document_loaders import DirectoryLoader, TextLoader, PyPDFLoader, CSVLoader

txt_loader = DirectoryLoader("resources", glob="**/*.txt", loader_cls=TextLoader)
pdf_loader = DirectoryLoader("resources", glob="**/*.pdf", loader_cls=PyPDFLoader)
csv_loader = DirectoryLoader("resources", glob="**/*.csv", loader_cls=CSVLoader)

docs = txt_loader.load() + pdf_loader.load() + csv_loader.load()
print(docs)
```
