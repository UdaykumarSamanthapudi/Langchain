```python
#Using Retriever to retrieve the documents
from langchain_community.vectorstores import  FAISS
from langchain_community.document_loaders import  TextLoader,PyPDFLoader
from langchain_text_splitters import  RecursiveCharacterTextSplitter
from langchain_huggingface import  HuggingFaceEmbeddings

loader=PyPDFLoader("resources/Java.pdf").load()

splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=50)
docs=splitter.split_documents(loader)

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_store=FAISS.from_documents(docs,embedding)
retriever=vector_store.as_retriever(search_kwargs={"k":3})

result=retriever.invoke("what is java?")
for r in result:
    print(r,"/n")
```

```python
mmr_retriever=vector_store.as_retriever(search_type="mmr",search_kwargs={"k":3})
res=mmr_retriever.invoke("what is java?")
for r in result:
    print(r,"/n")
```
